### 1차원 넘파이 배열
- elementwise 연산: 대응 되는 원소들끼리의 연산
- 벡터로써의 기능
- 벡터연산 가능

```python
arr1 = np.array([1,2,3])
arr1
```

```python
arr2 = np.array([4,5,6])
arr2
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
print(np.dot(arr1, arr2)) # inner product
```

### 2차원 넘파이 배열
- elementwise 연산
- 행렬로써의 기능
- 행렬 연산 가능

```python
mat1 = np.array([[1, 2], [3, 4]])
mat1
```

```python
mat2 = np.array([[5, 6], [7, 8]])
mat2
```

```python
print(mat1 + mat2)
print(mat1 - mat2)
print(mat1 * mat2) # element-wise
print(mat1 / mat2)
```

```python
# Load data
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Make class_names
class_names = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]

# Set size
plt.figure(figsize=(10,10))

# Get first 16 images and labels
images, labels = X_train[:16], y_train[:16].flatten()

# Print first 16 images and labels of train dataset of cifar10 
for i in range(16):
    # 4 by 4 image matrix for 16 images
    ax = plt.subplot(4, 4, i + 1)

    # Get 0 to 15 images as img
    img = images[i]

    # 이미지 출력
    plt.imshow(img)

    # set Tittle as matching class name: labels are non-negative integers
    plt.title(f"class: {class_names[labels[i]]}")

    # Do not show axis
    plt.axis("off")
```
